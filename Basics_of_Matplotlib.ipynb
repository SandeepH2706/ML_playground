 Precision-Recall Curve Plot


```
def plot_precision_recall(y_true, y_scores, model_name='Model'):
    """
    Plots the Precision-Recall curve.

    Parameters:
        y_true: Ground truth binary labels (0 or 1).
        y_scores: Predicted scores or probabilities for the positive class.
        model_name: Name to label the plot (optional).
    """
    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    avg_prec = average_precision_score(y_true, y_scores)

    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, label=f'{model_name} (AP = {avg_prec:.2f})', color='blue')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.grid(True)
    plt.legend()
    plt.show()
```
